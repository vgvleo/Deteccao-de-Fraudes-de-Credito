# 3. Pré-Processamento e Construção do Modelo

```python
#Separando os dados para treino e teste da máquina preditiva

x = df.drop(['Class'], axis = 1)
y = df['Class']
x_train, x_val, y_train, y_val = train_test_split(x,y, test_size = 0.3, random_state = 7)
```

```python
# Construindo modelo

model = RandomForestClassifier()
```

```python
# Treinando o modelo

model = model.fit(x_train,y_train)
```

```python
# Fazendo ingestão de novos dados (dados teste) para o modelo

y_predict = model.predict(x_val)
```

# 4. Avaliando Modelo Preditivo

```python
gabarito = pd.DataFrame({'Gabarito':y_val, 'Previsões_Modelo':y_predict})

gabarito
```

```python
# Avaliando modelo preditivo 

print('Classification metrics: \n',classification_report(y_val, y_predict))
```
